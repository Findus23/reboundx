```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig, ax = plt.subplots()
colors = ["red", "green", "blue", "cyan"]
for ID in [9]:
    for i, integ in enumerate(["implicit_midpoint", "euler", "naive", "none"]):
        data = np.loadtxt(integ+str(ID)+".txt", unpack=True)
        ax.plot(data[0], data[1], '.', color=colors[i], label=integ)
ax.set_xscale('log')
ax.set_yscale('log')
ax.legend(loc='lower left')
```
